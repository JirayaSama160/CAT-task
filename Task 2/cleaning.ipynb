```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
pd.set_option('display.max_columns', None)

df = pd.read_csv('train.csv')
df.head(10)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

df.drop(columns=['id'], inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

df.info()
```

## Brand

```python
df['brand'].unique()
```

```python
def replacing(replaces, col):
    df[col] = df[col].str.lower()
    for replace in replaces:
        df[col] = df[col].str.replace(replace, '')
    df[col] = df[col].str.strip()

replaces = ['i love', 'is the best']
replacing(replaces, 'brand')
df['brand'].unique()
```

```python
def grp_fillna(grp_list, col, mod):
    grp = df.groupby(grp_list)
    fill = ''
    if mod:
        fill = grp[col].transform(lambda x: np.nan if x.mode().empty else x.mode()[0])
    else:
        fill = grp[col].transform(lambda x: x.mean()//1)
    df[col] = df[col].fillna(fill)

grp_fillna(['vehiclemodel'], 'brand', True)
df['brand'].isna().sum()
```

```python
df['brand'] = df['brand'].fillna('other')
```

## year

```python
df['manufacturingyear'].unique()
```

```python
def extract_year(year):
    digits = r'\b\d{4}\b'
    match = re.search(digits, year)
    return match.group(0) if match else year

df['manufacturingyear'] = df['manufacturingyear'].replace({np.nan:'TEMPNAN'})
df['manufacturingyear'] = df['manufacturingyear'].apply(extract_year)
df['manufacturingyear'].unique()
```

```python
def calculate_year(year):
    if year.find('years old') != -1:
        year = year.replace('years old', '')
        return f'{2018 - int(year)}'
    return year

df['manufacturingyear'] = df['manufacturingyear'].str.replace('_', '')
df['manufacturingyear'] = df['manufacturingyear'].apply(calculate_year)
df['manufacturingyear'].unique()
```

```python
def correct_year(year):
    if year >= 3000:
        if 0 <= (year%100) <= 18:
            return year - 1000 
    return year

df['manufacturingyear'] = df['manufacturingyear'].replace({'TEMPNAN': np.nan})
df['manufacturingyear'] = df['manufacturingyear'].astype(float)
df['manufacturingyear'] = df['manufacturingyear'].apply(correct_year)

filt = (df['manufacturingyear'] >= 2024) | (df['manufacturingyear'] < 1950)
df.loc[filt, 'manufacturingyear'] = np.nan



print(df['manufacturingyear'].isna().sum())
df['manufacturingyear'].unique()
```

```python
group = df.groupby(['brand', 'vehiclemodel'])
group['manufacturingyear'].describe()
```

```python
grp_fillna(['brand', 'vehiclemodel'], 'manufacturingyear', True)
df = df.dropna(subset='manufacturingyear')
df.isna().sum()
```

## price

```python
df['price'].unique()
```

```python
df['price'].describe()
```

```python
grp_list = ['brand', 'vehiclemodel']
grp = df.groupby(grp_list)
grp['price'].describe()
```

```python
def calc_bounds(grp):
    q1 = grp['price'].quantile(0.25)
    q3 = grp['price'].quantile(0.75)
    iqr = q3-q1
    ub = q3 + 1.5*iqr
    lb = q1 - 1.5*iqr
    grp = (grp['price'] < lb) | (grp['price'] > ub)
    return grp

ff = grp.apply(calc_bounds).reset_index(drop=True)
ff.value_counts()
```

```python
df.reset_index(drop=True, inplace=True)
df['price_out'] = ff
filt = df['price_out'] == True

df.loc[filt, 'price'] = np.nan
df['price'].isna().sum()
```

```python
grp_list = ['brand', 'vehiclemodel']
while grp_list:
    grp_fillna(grp_list, 'price', False)
    grp_list.pop()
    
df['price'].isna().sum()
```
